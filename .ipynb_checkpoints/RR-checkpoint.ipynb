```python
import pandas as pd
import datetime as dt
import numpy as np
from tqdm.auto import tqdm
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

Я написал две функции для расчета RR, первая берет на вход дату и расчитывает retention по последующим дням, а также строит график с первого по 365 день, вторая функция берет дату начала периода и дату конца периода, а также день, для которого нужно посчитать RR, выводит график изменения по датам, а также топ10 дат с максимальным показателем retention на указанный день.
Данные хранятся в двух ДФ: 1. Дата регистрации (в виде timestamp) и id пользователя, 2. Дата авторизации в сервисе (в виде timestamp) и id пользователя.

```python
#функция, которая берет на вход дату и расчитывает retention по последующим дням, а также строит график с первого по 365 день
def retention_by_date(date):
    #загружаю данные, исходники выгрузил на я.диск, чтобы можно было по прямой ссылке
    df_time_reg =pd.read_csv('https://getfile.dokpub.com/yandex/get/https://disk.yandex.ru/d/djfKmxc-2gbpXA', sep=';')
    df_time_auth=pd.read_csv('https://getfile.dokpub.com/yandex/get/https://disk.yandex.ru/d/L8OYCyyrvNQlUA', sep=';')
    #меняю timestamp на date
    df_time_reg['reg_ts']=pd.to_datetime(df_time_reg.reg_ts, unit='s')
    df_time_auth['auth_ts']=pd.to_datetime(df_time_auth.auth_ts, unit='s')
    df_time_reg['reg_ts']=df_time_reg.reg_ts.dt.date
    df_time_auth['auth_ts']=df_time_auth.auth_ts.dt.date
    df_time_reg['reg_ts']=df_time_reg['reg_ts'].astype('datetime64')
    df_time_auth['auth_ts']=df_time_auth['auth_ts'].astype('datetime64')
    #создаю список с id пользователей зарегестрировавшихся в день, для которого считаю RR
    list_uid=df_time_reg.query('reg_ts==@date').uid.tolist()
    #создаю df для расчета RR,, отфильтровавшись по id пользователей из пред шага и сгруппировавшись по дням после даты регистрации и считаю кол-во id по каждому дню
    df_retent=df_time_auth.query('uid==@list_uid') \
        .groupby('auth_ts') \
        .agg({'uid':'nunique'}) \
        .rename(columns={'uid':'retention'})
    #считаю колонку retention 
    df_retent['retention']=(df_retent.retention/len(list_uid))*100
    df_retent=df_retent.reset_index().reset_index().rename(columns={'index':'Days'})
    #график(убрал из графика день 0, так как остальные дни, как правило не превышают 7-8%, из-за чего читаемость будет страдать)
    plt.figure(figsize=(20,10))
    sns.lineplot(data=df_retent[1:366], x='Days', y='retention')    
    #возвращаю табличку с RR на 1,2,3,7,30,90,180,365 день
    return df_retent[['Days','retention']].set_index('Days').loc[[1,2,3,7,30,90,180,365]]
```

```python
#функция, котрая берет дату начала периода и дату конца периода, а также день, для которого нужно посчитать RR, выводит график изменения по датам, а также топ10 дат с максимальным показателем retention на указанный день
def retention_by_day(date1,date2,td):
    #загружаю данные, исходники выгрузил на я.диск, чтобы можно было по прямой ссылке
    df_time_reg =pd.read_csv('https://getfile.dokpub.com/yandex/get/https://disk.yandex.ru/d/djfKmxc-2gbpXA', sep=';')
    df_time_auth=pd.read_csv('https://getfile.dokpub.com/yandex/get/https://disk.yandex.ru/d/L8OYCyyrvNQlUA', sep=';')
    #меняю timestamp на date    
    df_time_reg['reg_ts']=pd.to_datetime(df_time_reg.reg_ts, unit='s')
    df_time_auth['auth_ts']=pd.to_datetime(df_time_auth.auth_ts, unit='s')
    df_time_reg['reg_ts']=df_time_reg.reg_ts.dt.date
    df_time_auth['auth_ts']=df_time_auth.auth_ts.dt.date
    df_time_reg['reg_ts']=df_time_reg['reg_ts'].astype('datetime64')
    df_time_auth['auth_ts']=df_time_auth['auth_ts'].astype('datetime64')
    #создаю df с сэмплом данных, ограниченный внесенными датами
    df_time_reg1=df_time_reg.query('reg_ts>=@date1 and reg_ts<=@date2').reset_index()
    #создаю словарь с ключем датой и значением RR
    retent_td={}
    for i in tqdm(range(1, df_time_reg1.shape[0])):
        if df_time_reg1.reg_ts[i]!=df_time_reg1.reg_ts[i-1]:
            d1 = df_time_reg1.reg_ts[i]
            list_uid=df_time_reg1.query('reg_ts==@d1').uid.tolist()
            d=d1+td
            retent_td[d1]=df_time_auth.query('uid==@list_uid and auth_ts==@d').uid.nunique()/len(list_uid)
    #создаю df из словаря
    retention=pd.DataFrame(list(retent_td.items()), columns=['reg_date', 'retention'])
    retention=retention.set_index('reg_date')
    retention['retention']=retention.retention*100
    #график
    plt.figure(figsize=(18,10))
    sns.lineplot(data=retention)
    #возвращаю топ10 дат с максимальным показателем retention на указанный день
    return retention.sort_values('retention', ascending=False).head(10)
```

```python
print('Введите начальную дату для расчета показателя retention rate, в формате YYYY-MM-DD')
date=dt.datetime.strptime(input(), "%Y-%m-%d")
retention_by_date(date)
```

```python
print('Введите начальную дату периода для расчета показателя retention rate, в формате YYYY-MM-DD')
date1=dt.datetime.strptime(input(), "%Y-%m-%d")
print('Введите конечную дату периода для расчета показателя retention rate, в формате YYYY-MM-DD')
date2=dt.datetime.strptime(input(), "%Y-%m-%d")
print('Введите номер дня для расчета показателя retention rate')
td=dt.timedelta(days=int(input()))
retention_by_day(date1,date2,td)
```
